```python
if we have 2 catagorical features and we have to check whether null hypothesis is true or not
only on basis of these 2 catagorical features then we will use chi-square test.

    
if you have to check the acceptance or rejection of null hypothesis only on the basis of 1 feature
    having continuos variable data then you will use T-test.  (yani agr 1 feature ki base pr poora null hypothesis
depend kr rha ho , yani agar kafi features ho or null hypothese khy k "mean of weight is 30" to yha null hypothesis
  weight valy feature pr depend kr rha h )

    
if you have 1 numerical variable and 2 catagorical variables and in catagorical variables we have more than 2 catagories then
    we will use anove test. and if you have 1 numerical variable and 2 catagorical variables and in catagoricl variables we
have just 2 catagories then t-test will be used here
```

```python
Feature selection using the Chi-Square test is a statistical method used to determine the relationship between 
categorical features and the target variable.
    
You will use the Chi-Square test:
- When you have categorical data.
- To identify features that have a significant impact on the target variable.
- To improve model performance by removing irrelevant or redundant features.
```

```python
import seaborn as sns
df=sns.load_dataset('titanic')
```

```python
df.head()
```

```python
we will use catagorical features from this data set and then encode them to numerical and then
apply ch2 over it, it will give 2 va;ues, 1st one is f_score and second one is p_value
----> higher f_score and lwer p_value means that target feature is highly dependent on that fearure meaning thta independent feature is
very much important.
```

```python
selected_data=df.select_dtypes(include="object")
selected_data["target_survived"]=df["survived"]
selected_data.head()
```

```python
selected_data["embarked"].unique()
```

```python
selected_data["embark_town"].unique()
```

```python
selected_data["who"].unique()
```

```python
selected_data.isnull().sum()
```

```python
mode=selected_data["embarked"].mode()[0]
selected_data["embarked"]=selected_data["embarked"].fillna(mode)
selected_data["embarked"].isnull().sum()
```

```python
mode=selected_data["embark_town"].mode()[0]
selected_data["embark_town"]=selected_data["embark_town"].fillna(mode)
selected_data["embark_town"].isnull().sum()
```

```python
selected_data.isnull().sum()
```

```python
# encode that data:
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
en_data=selected_data[["sex","alive"]]
ohe=OneHotEncoder(drop="first")
encoded_data=ohe.fit_transform(en_data).toarray()
d=pd.DataFrame(encoded_data, columns=["sex_female","alive_yes"])
d
```

```python
selected_data[["sex_female","alive_yes"]]=d[["sex_female","alive_yes"]]
```

```python
selected_data=selected_data.drop(columns=["sex","alive"], axis=1)
```

```python
selected_data
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
en_data=selected_data["embarked"]
d=pd.Series(le.fit_transform(en_data))
selected_data["embarked_encoded"]=d
```

```python
selected_data=selected_data.drop("embarked", axis=1)
selected_data
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
en_data=selected_data["who"]
d=pd.Series(le.fit_transform(en_data))
selected_data["who_encoded"]=d
```

```python
selected_data=selected_data.drop("who", axis=1)
selected_data
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
en_data=selected_data["embark_town"]
d=pd.Series(le.fit_transform(en_data))
selected_data["embark_town_encoded"]=d
```

```python
selected_data=selected_data.drop("embark_town", axis=1)
selected_data
```

```python
from sklearn.model_selection import train_test_split
x=selected_data.drop("target_survived", axis=1)
y=selected_data["target_survived"]
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size=0.3)
```

```python
from sklearn.feature_selection import chi2
```

```python
f_p_value=chi2(x_train, y_train)
f_p_value
```

```python
data=pd.Series(f_p_value[0])
data.index=x_train.columns
data.sort_values(ascending=False) 
# big f_score and small p_score
```

```python
data=pd.Series(f_p_value[1])
data.index=x_train.columns
data.sort_values(ascending=False)     
```
