```python
import pandas as pd
```

```python
demo_data=pd.DataFrame({
    "A":[1,2,3,4,5],
    "B":[3,4,5,6,7],
    "C":[0,0,0,0,0],
    "D":[1,1,1,1,1]
})
demo_data
```

```python
Variance threshhold vo technique h jo k un columns ko delete krti h jin ka variance low hota h, 
yhan hm threshhold ki value pass krty hn, agar value 0 h to un sab columns ko drop kry gi jis ka
variance 0 ata, by defualt threshhold ki value 0 hi hoti, hm is ki value 0.1, 0.3 etc bhi rkh skty


--> variance threshold happens only for x features
```

```python
from sklearn.feature_selection import VarianceThreshold
vt=VarianceThreshold(threshold=0)
vt.fit(demo_data)
```

```python
vt.get_support()   #y btay ga k kon kon c features select huy hn, True un features 
# k liye jin k non_zero
# variance hn and False un k liye jin k zero variance hn
```

```python
sum(vt.get_support())  #sirf vo sum kr k do jo True valy hn
```

```python
demo_data.columns[vt.get_support()]   #sirf un k name do jo True valy hn
```

```python
non_0_var_cols=demo_data.columns[vt.get_support()] #yani vo columns do jo k True hn yani vo jin ki variance non_zero h
non_0_var_cols
```

```python
zero_var_cols=[]
for i in demo_data.columns:
    if i not in non_0_var_cols:
        zero_var_cols.append(i)
zero_var_cols
```

```python
demo_data=demo_data.drop(zero_var_cols, axis=1)
demo_data
```

```python
This feature selection algorithm looks only at the features (X), not the desired outputs (y),
and can thus be used for unsupervised learning.
# dividing data to testing and training done only for supervised learning
```

<h1> let's practice on original dataset </h1> 

```python
dataset=pd.read_csv("diabetes.csv")
dataset.head()
```

```python
x=dataset.drop("Outcome", axis=1)
y=dataset["Outcome"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3)
# TRAIN AND TEST SPLITTING IS DONE SO THAT WE CAN PREVENT FROM OVERFITTING
```

```python
from sklearn.feature_selection import VarianceThreshold
vt=VarianceThreshold(threshold=0)
vt.fit(x_train)
```

```python
vt.get_support()  #yani y bta rha h k saary features hi non_0_variance k hn
```

```python
sum(vt.get_support())  #yani total m s un ko sum kro jo True valy hn
```

```python
non_zero_var=x_train.columns[vt.get_support()]
non_zero_var
```

```python
zero_var=[]
for i in x_train.columns:
    if i not in non_zero_var:
        zero_var.append(i)
zero_var   
```

```python
x_train=x_train.drop(zero_var, axis=1)
x_train

# x-train s o columns niklaty vhi columns x-test s bhi nikaal deny hn withoir repeating the whole process again
```
