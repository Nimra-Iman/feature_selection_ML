```python
when our target variable is continues variable we use mutual information regarding regression and when
our target variable is discrete varaible we use mutual information regarding classification for
feature selection.                                                                                 
```

```python
import pandas as pd
from sklearn.datasets import load_diabetes
d=load_diabetes()
```

```python
dataset=pd.DataFrame(d.data, columns=d.feature_names)
dataset["target"]=d.target
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x=dataset.drop("target", axis=1)
y=dataset["target"]
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.feature_selection import mutual_info_regression
```

```python
mir=mutual_info_regression(x_train, y_train)
mir
```

```python
data=pd.Series(mir)
data.index=x_train.columns
data.sort_values(ascending=False)
```

```python
from sklearn.feature_selection import SelectPercentile
```

```python
# 20 percentile yani 20% data ap s neechy h, yani agr 
sp=SelectPercentile(mutual_info_regression, percentile=20)
# yani jo phly 20 percentiles m hn un ko select kro
sp.fit(x_train, y_train)
```

```python
selected_cols=x_train.columns[sp.get_support()]
selected_cols
```

```python
x_train=x_train[selected_cols]
x_train
```
